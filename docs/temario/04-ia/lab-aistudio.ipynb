## Bloque 2 — El system prompt como gramática prescriptiva

Escribir en el campo *System instructions* el siguiente conjunto de reglas:

```
Responde siempre en español.
No uses la voz pasiva en ningún caso.
No uses anglicismos ni préstamos léxicos.
Usa exclusivamente oraciones simples (sujeto + verbo + complementos directos).
El registro debe ser formal.
```

Lanzar después este prompt:

```
Explica qué es la inteligencia artificial y para qué se usa.
```

**Conexión conceptual:** un system prompt tiene la misma estructura lógica que una gramática formal: un conjunto de reglas que acotan el espacio de producciones válidas. La diferencia está en el mecanismo de seguimiento.
