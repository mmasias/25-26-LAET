# IA generativa — Sesión 1: Fundamentos

---

## ¿Por qué?

Las dos herramientas del bloque anterior tenían en común una limitación estructural: hay que saber exactamente qué se busca antes de buscarlo.

Con regex, hay que describir el patrón de caracteres. Con spaCy, hay que describir la función lingüística del token. En ambos casos, el análisis parte de una regla explícita definida de antemano.

La IA generativa opera con una lógica distinta: se le expresa una intención en lenguaje natural, y el sistema produce un resultado. No hace falta describir el patrón ni la función: basta con decir qué se necesita.

Esa diferencia no es cosmética. Cambia el tipo de preguntas que se pueden formular y el tipo de tareas que se pueden delegar.

---

## ¿Qué?

### Modelos de lenguaje de gran escala

Un **modelo de lenguaje de gran escala** (LLM, del inglés *large language model*) es un sistema entrenado sobre cantidades masivas de texto — páginas web, libros, artículos, código, conversaciones — para predecir cuál es el fragmento de texto más probable dado un contexto.

La intuición central es la del **autocompletado a escala masiva**.

Cuando se escribe en el buscador *El sinónimo de anglicismo es...*, el sistema no busca la respuesta en un diccionario: genera el texto más estadísticamente probable como continuación de esa frase, dado todo lo que ha procesado durante el entrenamiento. La diferencia con el autocompletado del teléfono móvil es la cantidad de datos de entrenamiento y la sofisticación de la arquitectura.

De esa naturaleza estadística surgen dos propiedades que conviene entender antes de usar estas herramientas en un contexto profesional.

---

### Temperatura

La temperatura es el parámetro que controla cuán estrictamente el modelo elige siempre el token más probable. Con **temperatura baja**, el modelo es predecible: elige casi siempre la continuación más probable. Con **temperatura alta**, elige opciones menos probables con más frecuencia, lo que produce respuestas más variadas y creativas, pero también menos precisas.

Las interfaces de uso general — Gemini, ChatGPT, Claude — no exponen este parámetro directamente, pero responden a instrucciones que lo afectan: pedir al modelo que sea *creativo* sube la temperatura efectiva; pedir que sea *preciso y conciso* la baja.

---

### Alucinaciones

Dado que el modelo genera siempre el texto más plausible estadísticamente, no tiene mecanismo para distinguir entre lo que sabe y lo que no sabe. Cuando se le pregunta por un dato que no aparecía en su corpus de entrenamiento — una referencia bibliográfica específica, una estadística concreta, un evento reciente — el modelo genera texto que *parece* una respuesta correcta: un título, un autor, un año, un DOI. Ese texto puede ser completamente inventado.

Este fenómeno se llama **alucinación**. No es un error ocasional: es una consecuencia estructural de cómo funcionan estos sistemas. Cualquier output que contenga datos verificables — citas, estadísticas, nombres propios, fechas — requiere verificación independiente.

---

## ¿Para qué?

Para el profesional de la lengua, los LLMs son útiles donde regex y PLN se detienen: tareas que requieren comprensión de contexto, intención o matiz.

| Tarea | Qué puede hacer regex / PLN | Qué añade un LLM |
|---|---|---|
| Traducción | Detectar términos en el texto de origen | Proponer traducciones con información de registro y contexto |
| Terminología | Extraer candidatos por frecuencia y categoría gramatical | Explicar diferencias semánticas entre términos candidatos |
| Post-edición | Detectar anglicismos por forma o categoría | Evaluar si una traducción automática es adecuada para un registro concreto |
| Creación de contenido | Generar variaciones mediante sustitución de patrones | Generar texto con restricciones de estilo, registro o audiencia |
| Investigación | Contar ocurrencias en corpus | Sintetizar información con indicación de procedencia |

La columna derecha no reemplaza a la izquierda. Son niveles de análisis complementarios: cada uno es la herramienta adecuada para preguntas distintas.

## ¿Cómo?

Los tres ejercicios siguientes usan directamente las interfaces web de los modelos disponibles: [Gemini](https://gemini.google.com), [ChatGPT](https://chatgpt.com), [Claude](https://claude.ai), [Copilot](https://copilot.microsoft.com). Los resultados pueden variar entre modelos: esa variación forma parte de lo que se observa.

Cada ejercicio indica el texto exacto que se envía al modelo. Las celdas de respuesta se completan con el texto del modelo tal como aparece, sin editar.

## Ejercicio 2: el efecto del contexto

El mismo modelo, la misma pregunta de fondo, dos versiones distintas. Se abren conversaciones independientes para cada una.

**Versión A** — sin contexto:

Se necesita una respuesta con terminología lingüística precisa y, si es posible, con referencia a la posición de la RAE o la Fundéu.*

## ¿Y ahora qué?

Los tres ejercicios muestran tres propiedades estructurales de los LLMs:

| Ejercicio | Propiedad observada | Consecuencia práctica |
|---|---|---|
| 1 | El modelo responde en lenguaje natural sobre conceptos del dominio | Puede usarse como primer punto de consulta, no como referencia definitiva |
| 2 | El contexto modifica la respuesta | El resultado depende de cómo se formula la petición |
| 3 | El modelo fabrica información verosímil | Cualquier dato concreto — cita, estadística, fecha — requiere verificación independiente |

La segunda propiedad es la que tiene mayor impacto práctico: si el resultado depende de la formulación, aprender a formular bien es una competencia directamente transferible al trabajo profesional. La sesión siguiente trabaja exactamente eso.
