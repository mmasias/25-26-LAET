# IA generativa — Sesión 2: Prompting básico

---

## ¿Por qué?

La sesión anterior mostró que el mismo modelo produce respuestas distintas según cómo se formula la petición. Ese resultado no es aleatorio: responde a una estructura. Los modelos de lenguaje procesan la entrada como un todo y ajustan su respuesta a cada elemento que detectan en ella.

Si el resultado depende de la formulación, construir bien una petición es una competencia. Se puede aprender y mejorar de forma sistemática, igual que se aprende a redactar un encargo de traducción o un pliego de condiciones.

A esa competencia se la llama *prompting*. El texto que se envía al modelo se llama *prompt*.

---

## ¿Qué?

### Anatomía de un prompt

Un prompt puede ser tan simple como una pregunta de una línea o tan elaborado como un documento con instrucciones detalladas. En cualquier caso, se puede descomponer en hasta cinco componentes:

| Componente | Qué define | Ejemplo |
|---|---|---|
| **Tarea** | Qué se le pide al modelo que produzca | *Redacta una ficha terminológica* |
| **Instrucciones** | Cómo debe producirlo: formato, extensión, restricciones | *en no más de 150 palabras, sin ejemplos de uso* |
| **Contexto** | Para qué, para quién, desde qué perspectiva | *para un glosario de traducción jurídica dirigido a traductores junior* |
| **Entrada** | El material concreto sobre el que trabaja el modelo | *Término: «due diligence»* |
| **Parámetros** | Configuraciones técnicas del modelo (temperatura, longitud máxima...) | *(raramente accesibles en interfaces web)* |

No todos los componentes son necesarios en todos los casos. Un prompt mínimo puede contener solo la tarea. Pero cuantos más componentes se especifican con precisión, más controlable es el output.

---

### El principio C³

Independientemente de los componentes que se usen, un prompt eficaz es:

- **Claro**: sin ambigüedades sobre qué se pide
- **Concreto**: con parámetros específicos (longitud, formato, audiencia, registro)
- **Conciso**: sin información irrelevante que desvíe la atención del modelo

| | Poco eficaz | Más eficaz |
|---|---|---|
| Claro | *Dime algo sobre traducción* | *Explica qué es la equivalencia dinámica en traducción* |
| Concreto | *Escribe sobre anglicismos* | *Escribe un párrafo de 80 palabras sobre anglicismos consolidados en el español jurídico* |
| Conciso | *Como sabrás, trabajo con textos muy técnicos y complicados, y me gustaría que...* | *Traduce el siguiente fragmento jurídico al español. Registro formal. Sin notas.* |

---

## ¿Para qué?

Un prompt bien construido permite:

- Obtener outputs en el formato exacto que se necesita (tabla, lista, párrafo, ficha)
- Controlar el registro y la audiencia del texto generado
- Delimitar el alcance de la respuesta
- Reutilizar la misma estructura para tareas similares, como una plantilla
- Reducir el número de iteraciones necesarias para llegar a un resultado utilizable

En contextos profesionales — traducción, localización, redacción técnica, investigación terminológica — la diferencia entre un prompt mínimo y uno bien construido puede ser la diferencia entre un output que requiere reescritura completa y uno que requiere solo revisión.

---

## ¿Cómo?

Los tres ejercicios trabajan la construcción progresiva de un prompt. El primero muestra el efecto de añadir componentes uno a uno. El segundo aísla el papel del contexto. El tercero propone construir un prompt para una tarea lingüística propia.

Conviene usar el mismo modelo en los tres ejercicios para que las diferencias sean atribuibles al prompt y no al modelo: [Gemini](https://gemini.google.com), [ChatGPT](https://chatgpt.com), [Claude](https://claude.ai), [Copilot](https://copilot.microsoft.com).

## ¿Y ahora qué?

### Lo que esta sesión ha establecido

| Componente | Función | Impacto en el output |
|---|---|---|
| Tarea | Define qué debe producir el modelo | Determina el tipo de respuesta |
| Instrucciones | Establece cómo debe producirlo | Controla formato, extensión y restricciones |
| Contexto | Indica para qué y para quién | Regula el registro, la densidad terminológica y el enfoque |
| Entrada | Proporciona el material de trabajo | Ancla la respuesta a un contenido específico |

Un prompt es, en ese sentido, un encargo profesional. Las mismas variables que determinan la calidad de un encargo de traducción — especificidad, contexto, restricciones, material de referencia — determinan la calidad de un prompt.

### La sesión siguiente

Las técnicas de prompting avanzado permiten ir más allá de la estructura de cuatro componentes: controlar el razonamiento del modelo paso a paso, proporcionar ejemplos que guíen el formato de la respuesta, encadenar peticiones para tareas complejas, o hacer que el modelo revise su propio output. Esas técnicas amplían el rango de tareas delegables y reducen la necesidad de post-edición extensiva.
