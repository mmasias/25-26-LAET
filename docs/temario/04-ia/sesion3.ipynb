# IA generativa — Sesión 3: Prompting avanzado

---

La sesión anterior estableció que un prompt bien construido tiene estructura (los cuatro componentes) y obedece al principio C³. Esa estructura es condición necesaria, pero no suficiente para tareas profesionales complejas.

Esta sesión trabaja tres técnicas que van más allá de la estructura básica: cómo mostrar al modelo lo que se espera en lugar de solo describirlo, cómo obligarlo a razonar de forma verificable y cómo reutilizar patrones de interacción que resuelven problemas recurrentes.

---

## Repaso: los cuatro componentes y el principio C³

| Componente | Qué define | Ejemplo |
|---|---|---|
| **Tarea** | Qué se le pide al modelo que produzca | *Redacta una ficha terminológica* |
| **Instrucciones** | Cómo debe producirlo: formato, extensión, restricciones | *en no más de 150 palabras, sin ejemplos de uso* |
| **Contexto** | Para qué, para quién, desde qué perspectiva | *para un glosario de traducción jurídica dirigido a traductores junior* |
| **Entrada** | El material concreto sobre el que trabaja el modelo | *Término: «due diligence»* |

Un prompt eficaz combina estos componentes siendo **claro** (sin ambigüedad sobre qué se pide), **concreto** (con parámetros específicos: longitud, formato, audiencia, registro) y **conciso** (sin información irrelevante que desvíe la atención del modelo). Estos son los tres principios del C³.

Las técnicas de esta sesión no sustituyen esa base: la amplían.

---

## ¿Por qué?

Tres limitaciones del prompting básico que aparecen en tareas profesionales reales.

**Describir el formato no es lo mismo que mostrarlo.** Se puede describir con detalle cómo debe ser una entrada terminológica (campo, definición, ejemplo de uso, nota de registro) y el modelo puede producir algo que respeta la descripción pero no el estilo ni la precisión que se espera. Mostrar dos entradas reales antes de pedir la tercera es más eficaz que cualquier descripción.

**Una opinión directa no es lo mismo que un razonamiento verificable.** Cuando se le pide al modelo que evalúe una traducción, produce una valoración que suena fundada. Pero esa valoración puede ser una generalización plausible, no un análisis real. Pedir al modelo que razone paso a paso hace visible el proceso: lo visible es verificable. La diferencia entre «primero terminología, luego registro, luego sintaxis» y una conclusión directa puede ser la diferencia entre un diagnóstico y una opinión.

**Algunos problemas de interacción se repiten.** Cómo conseguir que el modelo adopte el punto de vista de un experto específico. Cómo hacer que explique las razones de sus decisiones. Cómo conseguir que pida aclaraciones antes de producir output cuando la tarea es ambigua. Estos problemas tienen soluciones documentadas (patrones) que no hace falta reinventar cada vez.

---

## ¿Qué?

### Técnica 1: aprendizaje por ejemplos

Consiste en incluir en el prompt uno o varios ejemplos del tipo de output que se espera, antes de formular la petición real. El modelo extrae el patrón de los ejemplos y lo aplica al caso nuevo.

La variante sin ningún ejemplo se llama prompting sin ejemplos. Con uno o más ejemplos se habla de prompting con ejemplos. La diferencia no es terminológica: tiene consecuencias directas en la consistencia del formato y el estilo del output.

| Versión | Estructura del prompt |
|---|---|
| Sin ejemplos | Tarea + instrucciones + contexto + entrada |
| Con ejemplos | Ejemplo 1 - Ejemplo 2 - [Ejemplo N] - Entrada real |

Los ejemplos actúan como especificación implícita: transmiten lo que sería difícil o largo de describir con palabras. Para ver esta técnica aplicada a la enseñanza de un estilo de redacción, consultar el caso de uso [`05_redactor.md`](https://github.com/mmasias/ai-prompts/blob/main/documentos/casosDeUso/05_redactor.md) del repositorio ai-prompts.

---

### Técnica 2: razonamiento en cadena

Consiste en pedir al modelo que exponga su razonamiento paso a paso en lugar de producir directamente una conclusión. La forma más simple es añadir «piensa paso a paso» o «razona paso a paso»; la versión más controlada especifica explícitamente los pasos que se esperan.

El razonamiento en cadena no mejora el output porque el modelo «piense mejor»: lo mejora porque hace visible el proceso. Lo visible es verificable. Para tareas de evaluación (traducción, revisión de estilo, análisis de coherencia) la diferencia entre una respuesta directa y un razonamiento en cadena puede ser la diferencia entre una opinión y un diagnóstico.

---

### Técnica 3: patrones de interacción

Un patrón es una estructura de prompt que resuelve un problema recurrente de interacción con modelos de lenguaje. No son técnicas nuevas: son formas de estructurar los cuatro componentes conocidos de un modo que resuelve un problema específico de forma consistente.

| Patrón | Problema que resuelve | Estructura básica |
|---|---|---|
| **Persona** | El modelo responde desde una perspectiva genérica cuando se necesita una perspectiva experta | *«Actúa como [rol específico]. Tu criterio principal es [X].»* |
| **Reflexión** | El modelo produce resultados sin justificar sus decisiones | *«Por cada [sugerencia / cambio / decisión], explica el criterio que la justifica.»* |
| **Interacción invertida** | El modelo produce output antes de tener información suficiente | *«Hazme las preguntas necesarias para completar esta tarea. Cuando tengas suficiente información, produce el resultado.»* |

Los patrones se pueden combinar. Persona + Reflexión, por ejemplo, pide al modelo que adopte un rol experto y justifique sus decisiones desde ese rol. Para ver el patrón Persona aplicado a la corrección de estilo con criterios explícitos, consultar el caso de uso [`06_correctorEstilo.md`](https://github.com/mmasias/ai-prompts/blob/main/documentos/casosDeUso/06_correctorEstilo.md) del repositorio ai-prompts.

## ¿Para qué?

| Técnica | Tarea profesional | Por qué es útil |
|---|---|---|
| Aprendizaje por ejemplos | Generación de entradas terminológicas con formato de glosario | El formato exacto se transmite más eficazmente con 2 ejemplos que con instrucciones escritas |
| Aprendizaje por ejemplos | Producción de textos paralelos que replican el estilo de un texto de referencia | El modelo aprende el tono, el registro y la estructura del ejemplo antes de generar el texto nuevo |
| Razonamiento en cadena | Evaluación de la calidad de una traducción | Obliga al modelo a descomponer la evaluación por criterios, lo que hace el resultado verificable |
| Razonamiento en cadena | Análisis de coherencia terminológica en un documento | El modelo justifica cada observación, lo que facilita la revisión del análisis |
| Persona | Revisión de estilo con criterio especializado | El modelo adopta el punto de vista de un revisor con perfil definido, no el de un generador de texto genérico |
| Reflexión | Revisión de traducciones con justificación de cambios | Las justificaciones son verificables: el profesional puede aceptarlas o rebatirlas con criterio |
| Interacción invertida | Encargos de traducción con especificaciones incompletas | El modelo pide lo que le falta en lugar de asumir e inventar |

## ¿Y ahora qué?

| Técnica | Cuándo aplicarla |
|---|---|
| Aprendizaje por ejemplos | Cuando el formato o el estilo del output importan tanto como el contenido: glosarios, plantillas, textos paralelos |
| Razonamiento en cadena | Cuando se necesita un diagnóstico, no solo una opinión: evaluación de traducciones, análisis de coherencia, revisión crítica |
| Persona | Cuando el punto de vista desde el que se produce el output tiene impacto directo en la calidad: revisor, terminólogo, lector del texto de llegada |
| Reflexión | Cuando el output debe ser verificable o transferible: el criterio explícito permite aceptarlo o rebatirlo |
| Interacción invertida | Cuando el encargo tiene especificaciones incompletas: es preferible que el modelo pregunte a que invente |

Las técnicas de esta sesión no son procedimientos que se aplican mecánicamente: son herramientas que se combinan según la tarea. Un encargo complejo puede requerir aprendizaje por ejemplos para el formato, razonamiento en cadena para la evaluación del resultado y el patrón Reflexión para que el output sea auditable.

Las sesiones siguientes aplican estas técnicas a contextos específicos: traducción y localización asistida, creación de contenido lingüístico y evaluación crítica de outputs con criterio ético y profesional.
