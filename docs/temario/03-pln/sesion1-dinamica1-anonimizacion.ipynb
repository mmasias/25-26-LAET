# Dinámica 1: Anonimización automática de documentos

## Contexto profesional

Una de las aplicaciones más demandadas del NER en entornos profesionales es la **anonimización de documentos**: la sustitución sistemática de datos personales (nombres, organizaciones, lugares) por marcadores genéricos.

El Reglamento General de Protección de Datos (RGPD) obliga a anonimizar datos personales antes de compartir o publicar documentos sensibles. Hacerlo manualmente en un corpus de miles de páginas es inviable. El NER automatiza el primer paso; el criterio profesional valida el resultado.

## Objetivo de la dinámica

1. Aplicar NER para sustituir entidades por marcadores.
2. Evaluar qué escapó al modelo y qué fue anonimizado por error.
3. Reflexionar sobre los límites de la automatización en tareas con consecuencias legales.

```python
!python -m spacy download es_core_news_sm --quiet
import spacy

nlp = spacy.load("es_core_news_sm")
print("Modelo cargado.")
```

## La técnica

El procedimiento recorre las entidades detectadas en **orden inverso** (de derecha a izquierda en el texto) para sustituir cada tramo sin desplazar las posiciones de carácter de las entidades anteriores.

```python
marcadores = {
    "PER":  "[PERSONA]",
    "ORG":  "[ORGANIZACIÓN]",
    "LOC":  "[LUGAR]",
    "MISC": "[MISCELÁNEA]"
}

def anonimizar(texto):
    doc = nlp(texto)
    resultado = texto
    for entidad in reversed(doc.ents):
        marcador = marcadores.get(entidad.label_, f"[{entidad.label_}]")
        resultado = resultado[:entidad.start_char] + marcador + resultado[entidad.end_char:]
    return resultado
```

## Prueba con un texto de ejemplo

Se aplica la función sobre un fragmento de carta con datos personales ficticios.

```python
texto_original = """
Madrid, 3 de mayo de 2025

Estimada Dra. Carmen Iglesias:

Me dirijo a usted en representación de Laboratorios Ferrer S.A. para comunicarle
que el ensayo clínico coordinado por el Dr. Ramón Vega en el Hospital Gregorio Marañón
ha concluido satisfactoriamente. Los resultados han sido remitidos a la Agencia Española
de Medicamentos con sede en Madrid.

Atentamente,
Javier Morales
Director de Investigación
"""

print("--- TEXTO ORIGINAL ---")
print(texto_original)
print("--- TEXTO ANONIMIZADO ---")
print(anonimizar(texto_original))
```
