# Dinámica 2: Mapa de poder - ¿Quién tiene voz en la noticia?

## Contexto

Una noticia no es un espejo neutral de los hechos: es una selección de quién habla, quién es mencionado y quién está ausente. El análisis del discurso periodístico estudia estas relaciones como indicadores de poder, representación y sesgo editorial.

El NER proporciona una herramienta cuantitativa para comenzar ese análisis: en lugar de leer la noticia buscando nombres, el modelo los extrae automáticamente. Lo que antes requería una lectura minuciosa se convierte en una tabla ordenada por frecuencia.

## Objetivo de la dinámica

1. Extraer todas las personas mencionadas en una noticia.
2. Identificar cuáles tienen voz directa (son citadas) y cuáles solo son mencionadas.
3. Reflexionar sobre qué revela esa distribución acerca del enfoque editorial.

```python
!python -m spacy download es_core_news_sm --quiet
import spacy
import pandas as pd
from collections import Counter

nlp = spacy.load("es_core_news_sm")
print("Modelo cargado.")
```

## Paso 1: La noticia

Pega a continuación el cuerpo de una noticia de actualidad. Conviene que sea una noticia política, económica o social con varios actores mencionados.

```python
noticia = """
PEGA AQUÍ TU NOTICIA
"""

if noticia.strip() == "PEGA AQUÍ TU NOTICIA":
    print("Por favor, pega una noticia para continuar.")
else:
    doc = nlp(noticia)
    print(f"Noticia procesada: {len(list(doc.sents))} oraciones, {len(doc.ents)} entidades detectadas.")
```

## Paso 2: ¿Quién aparece y cuántas veces?

```python
personas = [ent.text.strip() for ent in doc.ents if ent.label_ == "PER"]
frecuencia = Counter(personas)

df_personas = pd.DataFrame(frecuencia.most_common(), columns=["Persona", "Menciones"])
print(f"Total de personas identificadas: {len(df_personas)}\n")
print(df_personas.to_string(index=False))
```

## Paso 3: ¿En qué contexto aparece cada persona?

Para cada persona del top 5, se muestran las oraciones en que aparece. Esto permite determinar si la persona habla en primera persona, si es citada, o si simplemente es mencionada.

```python
top_personas = [nombre for nombre, _ in frecuencia.most_common(5)]

for persona in top_personas:
    oraciones = [sent.text.strip() for sent in doc.sents if persona in sent.text]
    print(f"\n{'='*60}")
    print(f"  {persona}  ({frecuencia[persona]} menciones)")
    print(f"{'='*60}")
    for oracion in oraciones:
        print(f"  - {oracion}")
```

## Análisis

A partir de las oraciones de contexto, clasifica cada persona en una de estas categorías:

| Categoría | Descripción |
|---|---|

| **Agente** | Realiza acciones pero no tiene voz directa |
| **Referencia** | Es mencionada pero no actúa ni habla |
| **Ausente relevante** | No aparece pero su ausencia es significativa |
