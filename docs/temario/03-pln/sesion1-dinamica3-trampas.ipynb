# Dinámica 3: Las trampas lingüísticas

## Una forma de entender el modelo es romperlo

La mejor manera de comprender los límites de un sistema es construir los casos que lo hacen fallar. Esta dinámica invierte el flujo habitual: en lugar de usar el modelo para analizar textos, se usan los conocimientos lingüísticos para diseñar frases que lo engañen.

Cada trampa exitosa revela algo concreto sobre el mecanismo de decisión del modelo.

## Objetivo de la dinámica

1. Identificar los tipos de ambigüedad que confunden al modelo.
2. Construir frases diseñadas para provocar errores específicos.
3. Intercambiar las trampas con otro estudiante y comprobar si funcionan.

```python
!python -m spacy download es_core_news_sm --quiet
import spacy
from spacy import displacy

nlp = spacy.load("es_core_news_sm")
print("Modelo cargado.")
```

## Galería de trampas conocidas

En la sesión principal se vieron algunos ejemplos que inducen a error. Aquí se amplía la galería con análisis de por qué falla el modelo en cada caso.

```python
galeria = [
    ("Mañana viajaré a León.",                               "León: ciudad o nombre de animal"),
    ("El León es el rey de la selva.",                       "Mismo término, contexto diferente"),
    ("La señora Rosa compró rosas en el mercado.",           "Nombre propio idéntico al sustantivo común"),
    ("Apple presentó la nueva manzana de su logotipo.",      "Marca y fruta en el mismo contexto"),
    ("El joven Marcos llegó antes que el marco regulatorio.","Nombre propio vs. sustantivo técnico"),
    ("La victoria llevó a España a la final.",               "España: país o equipo deportivo"),
]

for frase, explicacion in galeria:
    print(f"\nTrampa: {frase}")
    print(f"Por qué puede fallar: {explicacion}")
    doc = nlp(frase)
    displacy.render(doc, style="ent", jupyter=True)
```

## Tipos de trampa

| Tipo | Mecanismo de confusión | Ejemplo |
|---|---|---|
| **Polisemia léxica** | La misma forma designa entidad y concepto común | *León* (ciudad / animal) |
| **Homonimia nombre/nombre** | Un nombre propio coincide con un sustantivo común | *Rosa*, *Marcos*, *Vera* |
| **Contexto contaminado** | La entidad real aparece junto a un uso figurado | *Apple presentó su manzana* |
| **Neologismo sin corpus** | El modelo nunca vio el término en su entrenamiento | *phishing*, *mansplaining* |
| **Mayúscula posicional** | Una palabra al inicio de frase recibe mayúscula no léxica | *La rosa de Jericó* |
| **Entidad implícita** | La entidad se sobreentiende sin nombrarse explícitamente | *El presidente anunció* |

## Tu turno: diseña tus trampas

Escribe al menos una frase por tipo. El objetivo es que el modelo clasifique mal la entidad, no la detecte, o detecte una donde no la hay.

Para cada frase, anota en la tupla el error esperado antes de ejecutar la celda.

```python
mis_trampas = [
    ("ESCRIBE AQUÍ TU PRIMERA TRAMPA",  "error esperado"),
    ("ESCRIBE AQUÍ TU SEGUNDA TRAMPA",  "error esperado"),
    ("ESCRIBE AQUÍ TU TERCERA TRAMPA",  "error esperado"),
]

for frase, esperado in mis_trampas:
    if "ESCRIBE AQUÍ" in frase:
        print(f"Pendiente: {frase}")
    else:
        print(f"\nTrampa: {frase}")
        print(f"Error esperado: {esperado}")
        doc = nlp(frase)
        displacy.render(doc, style="ent", jupyter=True)
```
